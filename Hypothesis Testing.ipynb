# Does the fuel type significantly affect the car price?

```python
import pandas as pd
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest

df = pd.read_csv('Cleaned_Data.csv')
print("Question: Does the Fuel type (Petrol vs Diesel) significantly affect the car price?\n")

group1 = df[df['fuel'] == 'Petrol']['price']
group2 = df[df['fuel'] == 'Diesel']['price']

n1 = len(group1)
n2 = len(group2)
mean1 = group1.mean()
mean2 = group2.mean()

print("H0 (Null): There is no significant difference in the average price between Petrol and Diesel cars. (μ_petrol = μ_diesel)")
print("H1 (Alternate):There is a significant difference in the average price between Petrol and Diesel cars. (μ_petrol ≠ μ_diesel)")
print()

if n1 < 30 or n2 < 30:
    print(f"Since one or both sample sizes are < 30 (n1={n1}, n2={n2}), we select the **Two-Sample T-Test**.")
    print()
    stat, p_val = stats.ttest_ind(group1, group2, equal_var=False)
else:
    print(f"Since both sample sizes are >= 30 (n1={n1}, n2={n2}), we select the **Two-Sample Z-Test**.")
    print()
    stat, p_val = ztest(group1, group2)
    
print(f"P-Value: {p_val}")

if p_val < 0.05:
    print("Decision: **Reject H0**.")
    
    if p_val < 0.001:
        strength = "Very Strong"
    elif p_val < 0.01:
        strength = "Strong"
    else:
        strength = "Moderate"
    
    print(f"Strength: {strength} evidence against the Null Hypothesis.")
    print()
else:
    print("Decision: **Failed to Reject H0** (Accept H0).")
    if p_val < 0.1:
        print("Strength: Weak evidence (Trending towards significance).")
    else:
        print("Strength: No significant evidence to suggest a difference.")
    print()

if p_val < 0.05:
    if mean1 < mean2:
        print("The analysis confirms that Diesel cars are priced significantly higher than Petrol cars.")
        print()
    else:
        print("The analysis confirms that Petrol cars are priced significantly higher than Diesel cars.")
        print()
else:
    print("There is no statistical evidence that fuel type affects price.")
    print()
```

```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv('Cleaned_Data.csv')
print("Question: Is there a significant correlation between KM Driven and Price?\n")

kms = df['km']
prices = df['price']
n = len(df)

print(f"Data Points: n={n}")
print("Variable 1: KM Driven")
print("Variable 2: Price")
print()

print("H0 (Null): There is no correlation between KM Driven and Price. (Correlation Coefficient r = 0)")
print("H1 (Alternate): There is a significant correlation between KM Driven and Price. (r ≠ 0)")
print()

print("Since we are checking the linear relationship between two continuous variables (KM and Price), we select the **Pearson Correlation Test**.")
print()
corr_coef, p_val = stats.pearsonr(kms, prices)

print(f"P-Value: {p_val}")
print(f"Correlation Coefficient: {corr_coef:.4f}")

if p_val < 0.05:
    print("Decision: **Reject H0**.")
    
    if p_val < 0.001:
        strength = "Very Strong"
    elif p_val < 0.01:
        strength = "Strong"
    else:
        strength = "Moderate"
    
    print(f"Strength: {strength} evidence against the Null Hypothesis.")
    print()
else:
    print("Decision: **Failed to Reject H0** (Accept H0).")
    if p_val < 0.1:
        print("Strength: Weak evidence (Trending towards significance).")
    else:
        print("Strength: No significant evidence to suggest a correlation.")
    print()

if p_val < 0.05:
    if corr_coef < 0:
        print("The analysis confirms a **Negative Correlation**.")
        print("Insight: As the Kilometers Driven increases, the Price significantly decreases. (Higher usage = Lower Value)")
    else:
        print("The analysis confirms a **Positive Correlation**.")
        print("Insight: As the Kilometers Driven increases, the Price increases (Unlikely for this data).")
    print()
else:
    print("There is no statistical evidence that KM driven affects price.")
    print()
```

```python
import pandas as pd
import scipy.stats as stats
from scipy.stats import chi2_contingency, f_oneway

df = pd.read_csv('Cleaned_Data.csv')



print("Question: Is the choice of Transmission dependent on Fuel Type?\n")


n = len(df)
print(f"Data Points: n={n}")
print("Variable 1: Fuel Type (Categorical)")
print("Variable 2: Transmission (Categorical)")
print()


print("H0 (Null): Fuel Type and Transmission are independent. (No relationship)")
print("H1 (Alternate): Fuel Type and Transmission are dependent. (Significant relationship)")
print()

print("Since we are checking the relationship between two categorical variables, we select the **Chi-Square Test of Independence**.")
print()

contingency_table = pd.crosstab(df['fuel'], df['transmission'])
chi2, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"P-Value: {p_val}")
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")

if p_val < 0.05:
    print("Decision: **Reject H0**.")
    
    if p_val < 0.001:
        strength = "Very Strong"
    elif p_val < 0.01:
        strength = "Strong"
    else:
        strength = "Moderate"
        
    print(f"Strength: {strength} evidence against the Null Hypothesis.")
    print("\nInsight: The choice of transmission is NOT random. Certain fuel types (e.g., Diesel vs Petrol) are statistically more likely to be paired with specific transmission types (Auto vs Manual).")
else:
    print("Decision: **Failed to Reject H0**.")
    print("\nInsight: Fuel type does not appear to influence the transmission type in this dataset.")
```

```python
print("Question: Does the Registration Location significantly affect the Car Price?\n")

top_5_regs = df['registration'].value_counts().head(5).index
subset_n = len(df[df['registration'].isin(top_5_regs)])
print(f"Data Points: n={subset_n} (Subset of Top 5 Locations)")
print("Variable 1: Registration Location (Categorical - Top 5)")
print("Variable 2: Price (Continuous)")
print(f"Groups Compared: {list(top_5_regs)}")
print()

print("H0 (Null): The mean price is the same across all selected locations.")
print("H1 (Alternate): At least one location has a significantly different mean price.")
print()

print("Since we are comparing the means of a numerical variable (Price) across more than two categorical groups (Locations), we select the **One-Way ANOVA Test**.")
print()

groups = [df[df['registration'] == r]['price'] for r in top_5_regs]
f_stat, p_val = f_oneway(*groups)

print(f"P-Value: {p_val}")
print(f"F-Statistic: {f_stat:.4f}")

if p_val < 0.05:
    print("Decision: **Reject H0**.")
    
    if p_val < 0.001:
        strength = "Very Strong"
    elif p_val < 0.01:
        strength = "Strong"
    else:
        strength = "Moderate"

    print(f"Strength: {strength} evidence against the Null Hypothesis.")
    print("\nInsight: Location matters. The state where a car is registered has a statistically significant impact on its resale price (likely due to tax/regulatory differences).")
else:
    print("Decision: **Failed to Reject H0**.")
    print("\nInsight: There is no statistical evidence that the registration location affects the price for these top 5 states.")
```
